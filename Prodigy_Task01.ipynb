```python
import pandas as pd

# Load the main dataset
df = pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_76034.csv', skiprows=4)

# Show first 5 rows
df.head()
```

```python
import matplotlib.pyplot as plt

# Sort and select top 10 populated countries
data = df[['Country Name', '2022']].dropna()
top10 = data.sort_values(by='2022', ascending=False).head(10)

# Create chart
plt.figure(figsize=(12,6))
plt.bar(top10['Country Name'], top10['2022'], color='lightgreen')
plt.title('Top 10 Most Populated Countries in 2022')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.tight_layout()

# ðŸ‘‰ Save the chart as an image (before showing)
plt.savefig('top10_population_chart.png')

# Show the chart
plt.show()

```
